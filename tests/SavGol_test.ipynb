```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(1, 'c:\\Users\\Vishal\\OneDrive - Texas Tech University\\lab-projects\\PeakSmoother\\PeakSmoother\\scripts')
import SavGol

import scienceplots
#plt.style.use('science')
plt.style.use(['science','no-latex'])
```

```python
data = pd.read_csv("../sample_data/4502.csv")
```

```python
intensity = data['Intensity'].values
length = len(intensity)
window_size = length // 200
smoothed_intensity = SavGol.savgol(intensity, window_size, 3)
```

```python
data['smoothed_intensity'] = smoothed_intensity
data['smoothed_intensity'] = data['smoothed_intensity'].apply(lambda x: 0 if x < 0 else x)
```

```python
plt.figure(figsize=(3.5, 2), dpi = 300)
plt.plot(data['RT'], data['Intensity'], label='Raw Data', color='black')
plt.plot(data['RT'], data['smoothed_intensity'], label='Smoothed Data', color='red', linewidth=1)
plt.xlabel('Retention Time (minutes)')
plt.ylabel('Intensity')
```
