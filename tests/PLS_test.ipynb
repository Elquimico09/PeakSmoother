```python
import sys
import os
sys.path.insert(1, 'C:\\Users\\visha\\OneDrive\\Documents\\GitRepos\\PeakSmoother\\scripts')
import PenalizedLeastSquare as PLS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("../sample_data/4502.csv")
```

```python
pls_smoothed_intensity = PLS.pls(data['RT'].values, data['Intensity'].values, lambda_param= 100)
```

```python
data['PLS'] = pls_smoothed_intensity

# Moving average on original intensity with window size 5
data['MA'] = data['Intensity'].rolling(window=5).mean()
```

```python
data_4502 = data
```

```python
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'no-latex'])
plt.rcParams['font.size'] = 16

# Plot the original intensity
plt.figure(figsize=(10, 12))

plt.subplot(3, 1, 1)
plt.plot(data_4502['RT'], data_4502['Intensity'], color = "blue")
plt.title('Original Intensity')
plt.xlim(0, 90)
plt.ylim(0, 1.1*max(data_4502['Intensity']))
plt.xlabel('Retention Time (min)')
plt.ylabel('Intensity')

# Plot the smoothed intensity
plt.subplot(3, 1, 2)
plt.plot(data_4502['RT'], data_4502['PLS'], color = "red")
plt.title('Penalized Least Square Smoothed Intensity')
plt.xlabel('Retention Time (min)')
plt.xlim(0, 90)
plt.ylim(0, 1.1*max(data_4502['Intensity']))
plt.ylabel('Intensity')
plt.tight_layout()

# Plot the moving average
plt.subplot(3, 1, 3)
plt.plot(data_4502['RT'], data_4502['MA'], color = "green")
plt.title('Moving Average')
plt.xlabel('Retention Time (min)', fontweight = "bold")
plt.xlim(0, 90)
plt.ylim(0, 1.1*max(data_4502['Intensity']))
plt.ylabel('Intensity')
plt.tight_layout()
plt.savefig("../figures/4502_PLS.png")
plt.show()
```
